## **2)— Analysing the Yearly Data Split By Clinic**

### **Context**

At the Vienna General Hospital, there were two maternity wards:

* **Clinic 1** → Staffed by male doctors and medical students
* **Clinic 2** → Staffed by female midwives

We now shift from monthly data to the **yearly dataset (`df_yearly`)** to compare the two clinics directly.

---

### **Challenge 1 — Compare Births and Deaths by Clinic**

We want to:

* Determine which clinic handled more births
* Check whether hospital activity increased over time
* Identify the highest number of deaths recorded in each clinic

---

### **Plot Total Yearly Births by Clinic**

We use Plotly and color by the `clinic` column:

```python
import plotly.express as px

line = px.line(
    df_yearly,
    x='year',
    y='births',
    color='clinic',
    title='Total Yearly Births by Clinic'
)

line.show()
```

#### **Observations**

* Births increased steadily over the years.
* **Clinic 1 consistently had more births** than Clinic 2.
* This suggests Clinic 1 was either larger or busier.

---

### **Plot Total Yearly Deaths by Clinic**

```python
line = px.line(
    df_yearly,
    x='year',
    y='deaths',
    color='clinic',
    title='Total Yearly Deaths by Clinic'
)

line.show()
```

#### **Observations**

* Clinic 1 also had more deaths.
* However, raw totals can be misleading because Clinic 1 handled more births.

To compare fairly, we must look at **proportions**, not totals.

---

## **Challenge 2 — Calculate the Proportion of Deaths**

To compare clinics fairly, we compute:

[
\text{Death Rate} = \frac{\text{Deaths}}{\text{Births}}
]

---

### **Add Percentage Column**

```python
df_yearly['pct_deaths'] = df_yearly.deaths / df_yearly.births
```

---

### **Average Death Rate per Clinic**

#### **Clinic 1**

```python
clinic_1 = df_yearly[df_yearly.clinic == 'clinic 1']

avg_c1 = clinic_1.deaths.sum() / clinic_1.births.sum() * 100

print(f'Average death rate in clinic 1 is {avg_c1:.3f}%')
```

Result:

* **~9.92%**

---

#### **Clinic 2**

```python
clinic_2 = df_yearly[df_yearly.clinic == 'clinic 2']

avg_c2 = clinic_2.deaths.sum() / clinic_2.births.sum() * 100

print(f'Average death rate in clinic 2 is {avg_c2:.3f}%')
```

Result:

* **~3.88%**

---

### **Visualise Proportion of Deaths**

```python
line = px.line(
    df_yearly,
    x='year',
    y='pct_deaths',
    color='clinic',
    title='Proportion of Yearly Deaths by Clinic'
)

line.show()
```

#### **Key Insight**

* In **1842**, death rates spiked dramatically:

  * Clinic 1 → ~16%
  * Clinic 2 → ~7.6%
* Clinic 2 consistently had a lower death rate.
* Clinic 1’s mortality rate was significantly higher across the years.

This is the central puzzle that frustrated Dr. Semmelweis.

---

## **The Narrative Turning Point**

Dr. Semmelweis tried multiple explanations:

* Delivery position differences → No effect
* Priest walking through ward ringing a bell → No effect

Then a breakthrough occurred:

A pathologist colleague died after cutting himself during an autopsy.
His symptoms matched those of women who died from childbed fever.

Semmelweis realized:

Doctors and students performing autopsies were likely transferring infectious material to mothers during delivery.

This insight laid the foundation for the **handwashing hypothesis**.

---

### **Core Learning Points**

* Always compare proportions when group sizes differ.
* Totals alone can mislead.
* Adding a calculated column (`pct_deaths`) allows fair comparison.
* Plotly’s `color` parameter makes grouped comparisons simple.
* Data often reveals patterns that narratives alone cannot.
