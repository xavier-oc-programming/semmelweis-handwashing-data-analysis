---

### **The Historical Context**

In the early 1800s:

* Germ theory did not yet exist.
* Illness was often attributed to “bad air” or superstition.
* Doctors began relying more on anatomy and autopsies.

Semmelweis noticed unusually high death rates in maternity wards and wanted to understand why. His investigation led to one of the earliest data-driven medical breakthroughs: handwashing drastically reduced maternal mortality.
