## **5) — Learning Points and Summary**

### What we discovered

By re-analysing Dr Semmelweis's original 1840s data we confirmed:

| Period | Avg. monthly death rate |
|--------|------------------------|
| Before handwashing (pre June 1846) | ~10.5% |
| After handwashing (June 1846 onward) | ~5.0% |
| Reduction | ~5.5 percentage points (~2× improvement) |

The t-test produced **p < 0.01**, meaning there is less than a 1% probability the improvement was due to chance alone.


### Key pandas / NumPy techniques used

| Technique | What it did |
|-----------|-------------|
| Column arithmetic | Computed `pct_deaths = deaths / births` |
| Boolean filtering | Split data into before/after handwashing subsets |
| `.rolling(6).mean()` | Smoothed month-to-month noise with a 6-month moving average |
| `np.where()` | Labelled rows as `'No'` / `'Yes'` for handwashing |
| `scipy.stats.ttest_ind()` | Quantified statistical significance of the drop |


### Visualisation techniques used

| Chart | Library | What it showed |
|-------|---------|----------------|
| Twin-axis line chart | Matplotlib | Monthly births vs deaths on shared time axis |

| Overlapping histogram | Plotly | Normalised frequency of monthly death rates |

### Historical reflection

Despite the statistical clarity of his findings, Dr Semmelweis was ridiculed by the medical establishment and died in a psychiatric institution in 1865 — two years before Louis Pasteur's germ theory provided the biological explanation that would have vindicated him.

This project is a reminder that **data alone does not change minds** — communication, framing, and receptive audiences matter just as much.
