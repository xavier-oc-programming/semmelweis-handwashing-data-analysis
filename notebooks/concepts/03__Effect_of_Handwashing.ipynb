## **3)— The Effect of Handwashing**

### Historical Context

In June 1846, Dr Semmelweis ordered doctors to wash their hands and instruments using a chlorine solution. Although germ theory was not yet established, he suspected that something connected autopsies to maternal deaths.

We now test whether the data supports the effectiveness of this intervention.

---

### Challenge 1 — Calculating the Death Rate Before and After Handwashing

#### Step 0 — Define the Intervention Date

Before splitting the dataset, we must define when handwashing began:

```python
import pandas as pd

handwashing_start = pd.to_datetime('1846-06-01')
```

This ensures we are comparing the correct time periods.

---

#### Step 1 — Create a Percentage Column

We compute the monthly proportion of deaths per birth:

```python
df_monthly['pct_deaths'] = df_monthly.deaths / df_monthly.births
```

This derived feature represents the monthly death rate.

---

#### Step 2 — Split the Data

```python
before_washing = df_monthly[df_monthly.date < handwashing_start]
after_washing  = df_monthly[df_monthly.date >= handwashing_start]
```

We now have:

* A pre-intervention dataset
* A post-intervention dataset

---

#### Step 3 — Compute Average Death Rates

```python
bw_rate = before_washing.deaths.sum() / before_washing.births.sum() * 100
aw_rate = after_washing.deaths.sum() / after_washing.births.sum() * 100

print(f'Average death rate BEFORE June 1846: {bw_rate:.4}%')
print(f'Average death rate AFTER June 1846: {aw_rate:.3}%')
```

#### Results

* **Before handwashing:** ~10.53%
* **After handwashing:** ~2.15%

This represents a dramatic reduction in maternal mortality.

---

### Challenge 2 — Rolling Average of the Death Rate

To reduce short-term noise, we compute a 6-month moving average before the intervention.

#### Step 1 — Set Date as Index

```python
roll_df = before_washing.set_index('date')
```

This preserves the time structure during rolling operations.

---

#### Step 2 — Compute Rolling Mean

```python
roll_df = roll_df.rolling(window=6).mean()
```

This smooths volatility and reveals the underlying mortality trend.

---

### Challenge 3 — Highlighting Subsections in a Line Chart

We now plot on a single axis:

* Death rate before handwashing
* 6-month moving average
* Death rate after handwashing

---

### Plot Structure

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.figure(figsize=(14,8), dpi=200)
plt.title('Percentage of Monthly Deaths over Time', fontsize=18)

years = mdates.YearLocator()
months = mdates.MonthLocator()
years_fmt = mdates.DateFormatter('%Y')

ax = plt.gca()
ax.xaxis.set_major_locator(years)
ax.xaxis.set_major_formatter(years_fmt)
ax.xaxis.set_minor_locator(months)

ax.set_xlim([df_monthly.date.min(), df_monthly.date.max()])
ax.set_ylabel('Percentage of Deaths', fontsize=18)

plt.grid(color='grey', linestyle='--')

ma_line, = plt.plot(
    roll_df.index,
    roll_df.pct_deaths,
    color='crimson',
    linewidth=3,
    linestyle='--',
    label='6m Moving Average'
)

bw_line, = plt.plot(
    before_washing.date,
    before_washing.pct_deaths,
    color='black',
    linewidth=1,
    linestyle='--',
    label='Before Handwashing'
)

aw_line, = plt.plot(
    after_washing.date,
    after_washing.pct_deaths,
    color='skyblue',
    linewidth=3,
    marker='o',
    label='After Handwashing'
)

plt.legend(handles=[ma_line, bw_line, aw_line], fontsize=14)
plt.show()
```

---

### Interpretation

The visual evidence is striking:

* Death rates were high and volatile before June 1846.
* The moving average confirms sustained elevated mortality.
* After handwashing began, the rate drops sharply and stabilizes.

This is a powerful early example of data-driven medical intervention.
